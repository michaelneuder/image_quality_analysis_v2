```python
import matplotlib.gridspec as gridspec
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# load images in
data_path = '/home/michaelneuder/iclr_images'
```

```python
# sample images 
plt.figure(figsize = (12,12))
gs1 = gridspec.GridSpec(3, 3)
gs1.update(wspace=0.03, hspace=0.03)

ax_dict = {}
for ii in range(9):
    ax_dict[ii] = plt.subplot(gs1[ii])
    ax_dict[ii].set_xticklabels([])
    ax_dict[ii].set_yticklabels([])
    ax_dict[ii].get_xaxis().set_visible(False)
    ax_dict[ii].get_yaxis().set_visible(False)

for ii in range(3):    
    temp_orig = mpimg.imread(os.path.join(data_path, '{}_orig.png'.format(ii)))
    temp_mse = mpimg.imread(os.path.join(data_path, '{}_kriz_mse_fc.png'.format(ii)))
    temp_ssim = mpimg.imread(os.path.join(data_path, '{}_kriz_ssim_fc.png'.format(ii)))
    
    ax_dict[3*ii].imshow(temp_orig, cmap='gray')
    ax_dict[3*ii+1].imshow(temp_ssim, cmap='gray')
    ax_dict[3*ii+2].imshow(temp_mse, cmap='gray')
    
ax_dict[0].set_title('original', size=20)
ax_dict[1].set_title('ssim', size=20)
ax_dict[2].set_title('mse', size=20)
# plt.savefig('iclr_sample.png')
plt.show()
```

```python
# sample images 
plt.figure(figsize = (9,9))
gs1 = gridspec.GridSpec(3, 3)
gs1.update(wspace=0.03, hspace=0.03)

ax_dict = {}
for ii in range(9):
    ax_dict[ii] = plt.subplot(gs1[ii])
    ax_dict[ii].set_xticklabels([])
    ax_dict[ii].set_yticklabels([])
    ax_dict[ii].get_xaxis().set_visible(False)
    ax_dict[ii].get_yaxis().set_visible(False)

_lambda = 1
for ii in range(3):   
    temp_orig = mpimg.imread(os.path.join(data_path, '{}_orig.png'.format(ii)))
    temp_mse = mpimg.imread(os.path.join(data_path, '{}_kriz_mse_fc.png'.format(ii)))
    temp_ssim = mpimg.imread(os.path.join(data_path, '{}_kriz_ssim_fc.png'.format(ii)))
    orig_mse = temp_orig + _lambda * temp_mse
    orig_ssim = temp_orig + _lambda * temp_ssim
    
    ax_dict[3*ii].imshow(temp_orig, cmap='gray')
    ax_dict[3*ii+1].imshow(orig_ssim, cmap='gray')
    ax_dict[3*ii+2].imshow(orig_mse, cmap='gray')
    

ax_dict[0].set_title('original', size=10)
ax_dict[1].set_title('original+({})ssim'.format(_lambda), size=10)
ax_dict[2].set_title('original+({})mse'.format(_lambda), size=10)
plt.savefig('iclr_sample_additions_lambda100.png'.format(int(_lambda*100)))
plt.show()
```
